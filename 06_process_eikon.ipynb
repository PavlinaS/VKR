```python
pl2.head()
```

```python
pl = pd.concat([pl1, pl2], ignore_index=True)
pl.head()
```

```python
    import math
    import datetime


    def xldate_to_datetime(xldatetime): #something like 43705.6158241088

      tempDate = datetime.datetime(1899, 12, 30)
      (days, portion) = math.modf(xldatetime)

      deltaDays = datetime.timedelta(days=days)
      #changing the variable name in the edit
      secs = int(24 * 60 * 60 * portion)
      detlaSeconds = datetime.timedelta(seconds=secs)
      TheTime = (tempDate + deltaDays + detlaSeconds )
      return TheTime.strftime("%Y-%m-%d %H:%M:%S")


xldate_to_datetime(43705.6158241088)
```

```python
for row in range(len(pl)):
    try:
        pl.Date[row] = xldate_to_datetime(pl.Date[row])
    except ValueError:
        continue
```

```python
pl['Date'] = pd.to_datetime(pl['Date'])
```

```python
pl.head()
```

```python
pl.to_csv(r'C:\Users\79627\Desktop\КР\данные\pl.csv', sep='|', index=False)
```
